Given a Pandas dataframe 'df' with columns 'gender' and 'age', how would you compute the average age for each gender?

```python
NUM_PEOPLE = 100
gender = ['M','F']
people_gender =  np.array(np.random.choice(gender, NUM_PEOPLE))
people_age = np.array(np.random.choice(range(100), NUM_PEOPLE))
```

```python
people_age
```

```python
df = pd.DataFrame({'gender': people_gender, 'age': people_age})
```

```python
df.groupby('gender')['age'].mean()
```
