# Pie Charts

A pie chart is a common univariate plot type that is used to depict relative frequencies for levels of a categorical variable. Frequencies in a pie chart are depicted as wedges drawn on a circle: the larger the angle or area, the more common the categorical value taken.

A  pie charts are a fairly limited plot type in the range of scenarios where they can be used, and it is easy for chart makers to try and spice up pie charts in a way that makes them more difficult to read. 

To use a pie chart:
- Make sure that your interest is in relative frequencies
- Limit the number of slices plotted. A pie chart works best with two or three slices, though it's also possible to plot with four or five slices as long as the wedge sizes can be distinguished
- Plot the data systematically. One typical method of plotting a pie chart is to start from the top of the circle, then plot each categorical level clockwise from most frequent to least frequent. If you have three categories and are interested in the comparison of two of them, a common plotting method is to place the two categories of interest on either side of the 12 o'clock direction, with the third category filling in the remaining space at the bottom.

If these guidelines cannot be met, then you should probably make use of a bar chart instead. A bar chart is a safer choice in general

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# this allow to use matplotlib in the notebook
%matplotlib inline 
```

```python
# importing data
pokemon_data = pd.read_csv('./data/pokemon.csv')
pkmn_type = pokemon_data.melt(id_vars=['id', 'species'], value_vars=['type_1', 'type_2'], var_name='type_level', value_name='type').dropna()
```

To create a pie
- "startangle = 90" and "counterclock = False" to fullfil the previous bullet point
- set axis to square so that the scaling of the plot is equal on both the x- and y-axes. Without this call, the pie could end up looking oval-shaped, rather than a circle. ([doc](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axis.html))

```python
sorted_counts = pkmn_type['type'].value_counts()
subset_counts = sorted_counts[0:4]

others = pd.Series({'others': sorted_counts[5:].sum()})
subset_counts = subset_counts.append(others)
```

```python
plt.figure(figsize=(16, 6))
plt.pie(subset_counts, labels = subset_counts.index, startangle = 90, counterclock = False)
plt.axis('square');
```

A sub-category of the pie chart family is the donut chart. Wherever you can use a pie chart, you might as well use a donut chart because we don't read pie charts by angle of each wedge.

It's possible to create this kind of chart by using the [wedgeprops](https://matplotlib.org/api/_as_gen/matplotlib.patches.Wedge.html)

```python
plt.figure(figsize=(16, 6))
plt.pie(subset_counts, labels = subset_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.5})
plt.axis('square');
```
